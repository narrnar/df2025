```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Filter for South Bay, Class A, large leases only
southbay = leases[
    (leases['market'] == 'Los Angeles') &
    (leases['internal_submarket'] == 'South Bay') &
    (leases['internal_class'] == 'A') &
    (leases['leasedSF'].astype(float) >= 2000)  # convert in case column is string type
]

# 2. Create a quarter identifier (e.g., '2019 Q3')
# Make sure 'year' and 'quarter' are numeric and present
southbay['year_qtr'] = southbay['year'].astype(str) + ' Q' + southbay['quarter'].astype(str)

# 3. Group by quarter to get mean vacancy and rent
# We'll use 'availability_proportion' and 'overall_rent'
quarterly = southbay.groupby('year_qtr').agg(
    mean_vacancy=('availability_proportion', 'mean'),
    mean_rent=('overall_rent', 'mean')
).reset_index()

# 4. Lead rent 2 quarters ahead
quarterly['rent_lead2'] = quarterly['mean_rent'].shift(-2)

# 5. Calculate 2-quarter forward rent growth %
quarterly['rent_growth_2q'] = (quarterly['rent_lead2'] - quarterly['mean_rent']) / quarterly['mean_rent'] * 100

# 6. Bin the vacancy into labeled ranges
bins = [0, 0.10, 0.125, 0.15, 0.175, 0.20, np.inf]
labels = ['0-10%', '10-12.5%', '12.5-15%', '15-17.5%', '17.5-20%', '20%+']
quarterly['vacancy_bin'] = pd.cut(quarterly['mean_vacancy'], bins=bins, labels=labels)

# 7. Create heatmap pivot table: rows = bins, cols = quarters, values = rent growth
pivot_table = quarterly.pivot(index='vacancy_bin', columns='year_qtr', values='rent_growth_2q')

# 8. Plot
plt.figure(figsize=(14, 6))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="RdYlGn", center=0, linewidths=.5)
plt.title('South Bay Class A: Future Rent Growth (%) by Vacancy Bin and Quarter')
plt.ylabel('Vacancy Rate Bin')
plt.xlabel('Quarter')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import statsmodels.formula.api as smf

# Make sure we drop NA from rent_growth
seg_data = quarterly.dropna(subset=['rent_growth_2q'])

# Create a variable for "below 15%" (the knot)
seg_data['vacancy_below_15'] = np.where(seg_data['mean_vacancy'] < 0.15, seg_data['mean_vacancy'] - 0.15, 0)
seg_data['vacancy_above_15'] = np.where(seg_data['mean_vacancy'] >= 0.15, seg_data['mean_vacancy'] - 0.15, 0)

# Fit the piecewise model
model = smf.ols('rent_growth_2q ~ vacancy_below_15 + vacancy_above_15', data=seg_data).fit()
print(model.summary())
```

```python
# Define binary outcome: high growth = 1 if  5%
seg_data['high_rent_growth'] = (seg_data['rent_growth_2q'] >= 5).astype(int)

# Fit logistic regression
logit_model = smf.logit('high_rent_growth ~ mean_vacancy', data=seg_data).fit()
print(logit_model.summary())

seg_data['predicted_prob'] = logit_model.predict(seg_data)

plt.figure(figsize=(8, 4))
sns.scatterplot(x='mean_vacancy', y='high_rent_growth', data=seg_data, label='Actual')
sns.lineplot(x='mean_vacancy', y='predicted_prob', data=seg_data, color='orange', label='Predicted Prob')
plt.axvline(0.15, linestyle='--', color='red', label='15% Threshold')
plt.title('Prob. of Rent Growth 5% by Vacancy Rate (South Bay)')
plt.ylabel('Probability')
plt.xlabel('Mean Vacancy Rate')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# When South Bay vacancy is < 15 %, does Class-A effective rent rise within the next two quarters at  2 the pace it rises when vacancy is  15 %?
# ================================================================
#  South Bay trigger test  revised column targets
# ================================================================

import pandas as pd, numpy as np, re
from pathlib import Path
from scipy import stats
import matplotlib.pyplot as plt

# ----------------------------------------------------------------
# 1  Load
# ----------------------------------------------------------------
root   = Path("/Users/darensivam/Desktop/UCLA/dataFest2025/2025 Data Files")
leases = pd.read_csv(root / "Leases.csv")
occup  = pd.read_csv(root / "Major Market Occupancy Data-revised.csv")

# ----------------------------------------------------------------
# 2  Helpers
# ----------------------------------------------------------------
def make_quarter(y, q):
    q = q.astype(str).str.replace(r"^Q", "", regex=True)   # strip leading Q
    return y.astype(str) + "Q" + q

def fuzzy(series, needle):
    """case-insensitive contains test, ignoring dashes/spaces"""
    needle = re.sub(r"[\- ]+", " ", needle.lower()).strip()
    return (series.astype(str).str.lower()
            .str.replace(r"[\- ]+", " ", regex=True)
            .str.contains(needle, regex=False))

# ----------------------------------------------------------------
# 3  Filters
# ----------------------------------------------------------------
TARGET = "south bay"

# --- market / sub-market / cluster ------------------------------
mkt_mask = (
      fuzzy(leases["internal_market_cluster"], TARGET)
    | fuzzy(leases["internal_submarket"],      TARGET)
    | fuzzy(leases["market"],                  TARGET)
)

# --- class = 'A' -------------------------------------------------
class_mask = leases["internal_class"].str.strip().str.upper().eq("A")

# --- size  10 k sf ---------------------------------------------
size_mask  = leases["leasedSF"].ge(10_000)

leases_sb = (
    leases.loc[mkt_mask & class_mask & size_mask]
          .assign(Quarter = make_quarter(leases["year"], leases["quarter"]))
)

print(f"Leases  after filters: {len(leases_sb):5d} rows")

# ----------------------------------------------------------------
# 4  Rent series (median $/sf per quarter)
# ----------------------------------------------------------------
rent_q = (leases_sb.groupby("Quarter", as_index=False)
                    .agg(eff_rent_psf=("internal_class_rent", "median")))

# ----------------------------------------------------------------
# 5  Vacancy series   fuzzy-match occupancy table too
# ----------------------------------------------------------------
occ_mask = fuzzy(occup["market"], TARGET)
vac_sb = (
    occup.loc[occ_mask]
         .assign(
             Quarter = make_quarter(occup["year"], occup["quarter"]),
             Vacancy = (1 - occup.loc[occ_mask, "avg_occupancy_proportion"]) * 100
         )[["Quarter", "Vacancy"]]
)

print(f"Occupancy rows kept : {len(vac_sb):5d}")

# ----------------------------------------------------------------
# 6  Merge and 2-quarter forward rent growth
# ----------------------------------------------------------------
df = (rent_q.merge(vac_sb, on="Quarter", how="inner")
           .sort_values("Quarter")
           .reset_index(drop=True))

print(f"Merged rent-vac rows: {len(df):5d}")

if df.empty:
    raise ValueError("No overlapping quarters  check Quarter strings.")

df["rent_fwd2_pct"] = df["eff_rent_psf"].pct_change(periods=2).shift(-2)

low   = df[df["Vacancy"]  < 15]
high  = df[df["Vacancy"] >= 15]

print(f"Low-vac rows (<15%) : {len(low):5d}")
print(f"High-vac rows(15%) : {len(high):5d}")

# ----------------------------------------------------------------
# 7  Stats
# ----------------------------------------------------------------
g_low, g_high = low["rent_fwd2_pct"].mean(), high["rent_fwd2_pct"].mean()
ratio = g_low / g_high if pd.notna(g_high) and g_high != 0 else np.nan

print(f"\nAvg 2-Q growth  <15%: {g_low: .2%}")
print(f"Avg 2-Q growth 15%: {g_high: .2%}")
print(f"Ratio (need 2)     : {ratio: .2f}")

t, p = stats.ttest_ind(
        low["rent_fwd2_pct"].dropna(),
        high["rent_fwd2_pct"].dropna(),
        equal_var=False)
print(f"Welch t-stat = {t: .2f},  p-value = {p: .3f}")

# ----------------------------------------------------------------
# 8  Visuals
# ----------------------------------------------------------------
plt.figure(figsize=(9, 4))
plt.plot(df["Quarter"], df["eff_rent_psf"], marker="o")
plt.xticks(rotation=45, ha="right")
plt.ylabel("$/sf")
plt.title("LA South Bay  Class-A Effective Rent by Quarter")
plt.tight_layout()

plt.figure(figsize=(6, 4))
plt.scatter(df["Vacancy"], df["rent_fwd2_pct"])
plt.axvline(15, color="grey", ls="--")
plt.xlabel("Vacancy % (current qtr)")
plt.ylabel("2-Q Fwd Rent Growth %")
plt.title("Rent Growth vs. 15 % Vacancy Threshold")
plt.tight_layout()
plt.show()
```

```python
# --- For occupancy table check ---
# sb_occ = occup.loc[fuzzy(occup["market"], "south bay")]   # same mask as before
# sb_occ.head(3)[["ending_occupancy_proportion",
#                 "starting_occupancy_proportion",
#                 "avg_occupancy_proportion"]]

# ================================================================
# South Bay   best vacancy breakpoint search (supply-side availability, not desk-occupancy)
# ================================================================

import pandas as pd, numpy as np, re
from pathlib import Path
from scipy import stats
import matplotlib.pyplot as plt

# ----------------------------------------------------------------
# Load tables
# ----------------------------------------------------------------
root   = Path("/Users/darensivam/Desktop/UCLA/dataFest2025/2025 Data Files")
leases = pd.read_csv(root / "Leases.csv")
pa     = pd.read_csv(root / "Price and Availability Data.csv")

# ----------------------------------------------------------------
# Helpers
# ----------------------------------------------------------------
def make_quarter(y, q):
    q = q.astype(str).str.replace(r"^Q", "", regex=True)  # strip leading 'Q'
    return y.astype(str) + "Q" + q

def fuzzy(series, needle):
    needle = re.sub(r"[\- ]+", " ", needle.lower()).strip()
    return (series.astype(str).str.lower()
            .str.replace(r"[\- ]+", " ", regex=True)
            .str.contains(needle, regex=False))

# ----------------------------------------------------------------
# RENT series  (South Bay  Class A  10 k sf)
# ----------------------------------------------------------------
TARGET = "south bay"

lease_mask = (
      fuzzy(leases["internal_market_cluster"], TARGET)
    | fuzzy(leases["internal_submarket"],      TARGET)
    | fuzzy(leases["market"],                  TARGET)
) & leases["internal_class"].str.strip().str.upper().eq("A") \
  & leases["leasedSF"].ge(10_000)

rent_q = (leases.loc[lease_mask]
                 .assign(Quarter = make_quarter(leases["year"], leases["quarter"]))
                 .groupby("Quarter", as_index=False)
                 .agg(eff_rent_psf=("internal_class_rent", "median")))

print(f"Rent quarters: {len(rent_q)}")

# ----------------------------------------------------------------
# VACANCY (availability) series   Price & Availability table
# ----------------------------------------------------------------
pa_mask = fuzzy(pa["market"], TARGET) \
          & pa["internal_class"].str.strip().str.upper().eq("A")

avail_col = "availability_proportion"        # present in the column list you sent

vac_q = (pa.loc[pa_mask]
           .assign(
               Quarter = make_quarter(pa["year"], pa["quarter"]),
               Vacancy = pa.loc[pa_mask, avail_col] * 100     # convert 0-1  %
           )
           .groupby("Quarter", as_index=False)
           .agg(Vacancy=("Vacancy", "mean")))

print(f"Vacancy quarters: {len(vac_q)}")

# ----------------------------------------------------------------
# Merge & two-quarter forward rent growth
# ----------------------------------------------------------------
df = (rent_q.merge(vac_q, on="Quarter", how="inner")
            .sort_values("Quarter")
            .reset_index(drop=True))

df["rent_fwd2_pct"] = df["eff_rent_psf"].pct_change(periods=2).shift(-2)
print(f"Merged quarters (rent  vacancy): {len(df)}")

# ----------------------------------------------------------------
# Find the vacancy % that best splits slow vs. fast growth
# ----------------------------------------------------------------
def best_split(data, min_n=3):
    xs = data["Vacancy"].values
    ys = data["rent_fwd2_pct"].values
    thresholds = np.unique(xs)[1:-1]          # skip min & max
    best = {"thr": None, "t": -np.inf}
    for thr in thresholds:
        low  = ys[xs <  thr]
        high = ys[xs >= thr]
        if len(low) < min_n or len(high) < min_n:
            continue
        t_val, _ = stats.ttest_ind(low, high, equal_var=False)
        if abs(t_val) > best["t"]:
            best.update({"thr": thr, "t": abs(t_val),
                         "low_n": len(low), "high_n": len(high),
                         "low_mean": low.mean(), "high_mean": high.mean()})
    return best

split = best_split(df.dropna(subset=["rent_fwd2_pct"]))

if split["thr"] is None:
    raise ValueError("Not enough data on both sides of any split.")

print("\n=== Optimal availability breakpoint ===")
print(f"Threshold           : {split['thr']:.1f} %")
print(f"Obs < threshold     : {split['low_n']}   |   mean growth = {split['low_mean']:.2%}")
print(f"Obs  threshold     : {split['high_n']}  |   mean growth = {split['high_mean']:.2%}")
print(f"|Welch t| statistic  : {split['t']:.2f}")

# ----------------------------------------------------------------
# Visual check
# ----------------------------------------------------------------
plt.figure(figsize=(6,4))
plt.scatter(df["Vacancy"], df["rent_fwd2_pct"])
plt.axvline(split["thr"], color="red", ls="--", label=f"Best Split = {split['thr']:.1f}%")
plt.xlabel("Availability / Vacancy % (current quarter)")
plt.ylabel("2-Q Forward Rent Growth %")
plt.title("South Bay  Class A   Vacancy vs. Forward Rent Growth")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ------------------------------------------------------------------
#  Helper - Run full breakpoint analysis for one micro-market
def analyse_market(target, plot=True, min_size=2_000, class_letter="A"):
    """Return dict with breakpoint stats and (optionally) show plot."""
    # --- RENT ---
    lease_mask = (
          fuzzy(_leases["internal_market_cluster"], target)
        | fuzzy(_leases["internal_submarket"],      target)
        | fuzzy(_leases["market"],                  target)
    ) & _leases["internal_class"].str.upper().eq(class_letter) \
      & _leases["leasedSF"].ge(min_size)

    rent_q = (_leases.loc[lease_mask]
                     .assign(Quarter = make_quarter(_leases["year"], _leases["quarter"]))
                     .groupby("Quarter", as_index=False)
                     .agg(eff_rent_psf=("internal_class_rent", "median")))

    # --- VACANCY ---
    pa_mask = fuzzy(_pa["market"], target) \
              & _pa["internal_class"].str.upper().eq(class_letter)

    vac_q = (_pa.loc[pa_mask]
               .assign(Quarter = make_quarter(_pa["year"], _pa["quarter"]),
                       Vacancy  = _pa.loc[pa_mask, "availability_proportion"] * 100)
               .groupby("Quarter", as_index=False)
               .agg(Vacancy=("Vacancy", "mean")))

    merged = (rent_q.merge(vac_q, on="Quarter", how="inner")
                     .sort_values("Quarter")
                     .reset_index(drop=True))
    merged["rent_fwd2_pct"] = merged["eff_rent_psf"].pct_change(periods=2).shift(-2)
    merged = merged.dropna(subset=["rent_fwd2_pct"])

    # --- best split ---
    res = best_split(merged, min_n=3)      # uses function from earlier cell
    res.update({
        "market" : target.title(),
        "rent_q" : len(rent_q),
        "vac_q"  : len(vac_q),
        "merged_q": len(merged)
    })

    if plot:
        plt.figure(figsize=(6,4))
        plt.scatter(merged["Vacancy"], merged["rent_fwd2_pct"])
        plt.axvline(res["thr"], color="red", ls="--", label=f'best split = {res["thr"]:.1f}%')
        plt.title(f'{target.title()}   Class A\nVacancy vs 2-Q Rent Growth')
        plt.xlabel("Availability % (current qtr)")
        plt.ylabel("2-Q Fwd Rent Growth %")
        plt.legend()
        plt.tight_layout()
        plt.show()

    print(f"{target.title():<12}  |  thr {res['thr']:.1f}%  "
          f"|  n_low {res['low_n']:2d}   {res['low_mean']:.2%}  "
          f"|  n_high {res['high_n']:2d}   {res['high_mean']:.2%}  "
          f"|  |t| {res['t']:.2f}")
    return res

# ------------------------------------------------------------------
#  Helper - batch-compare several markets
def compare_markets(markets):
    records = [analyse_market(m, plot=False) for m in markets]
    cols = ["market","thr","low_mean","high_mean","low_n","high_n","t"]
    df_cmp = pd.DataFrame(records)[cols]
    return df_cmp.sort_values("thr").reset_index(drop=True)

# ------------------------------------------------------------------
MIN_SIZE = 2_000 
class_letter = "A"

def best_split(data, *, min_n=3, x="Vacancy", y="rent_fwd2_pct"):
    xs, ys = data[x].values, data[y].values
    thresholds = np.unique(xs)[1:-1]   
    best = {"thr": None, "t": -np.inf,
            "low_n": 0, "high_n": 0,
            "low_mean": np.nan, "high_mean": np.nan}
    for thr in thresholds:
        low  = ys[xs <  thr]
        high = ys[xs >= thr]
        if len(low) < min_n or len(high) < min_n:
            continue
        t_val, _ = stats.ttest_ind(low, high, equal_var=False)
        if abs(t_val) > best["t"]:
            best.update({"thr": thr, "t": abs(t_val),
                         "low_n": len(low),  "high_n": len(high),
                         "low_mean": low.mean(), "high_mean": high.mean()})
    return best

def analyse_market(target, *, min_size=MIN_SIZE, min_n=3, plot=True):
    """Run breakpoint search for one micro-market."""
    tgt = target.lower()

    lease_mask = (
          fuzzy(_leases["internal_market_cluster"], tgt)
        | fuzzy(_leases["internal_submarket"],      tgt)
        | fuzzy(_leases["market"],                  tgt)
    ) & _leases["internal_class"].str.upper().eq(class_letter) \
      & _leases["leasedSF"].ge(min_size)

    rent_q = (_leases.loc[lease_mask]
                     .assign(Quarter = make_quarter(_leases["year"], _leases["quarter"]))
                     .groupby("Quarter", as_index=False)
                     .agg(eff_rent_psf=("internal_class_rent", "median")))

    pa_mask = fuzzy(_pa["market"], tgt) \
              & _pa["internal_class"].str.upper().eq(class_letter)

    vac_q = (_pa.loc[pa_mask]
               .assign(Quarter = make_quarter(_pa["year"], _pa["quarter"]),
                       Vacancy  = _pa.loc[pa_mask, "availability_proportion"] * 100)
               .groupby("Quarter", as_index=False)
               .agg(Vacancy=("Vacancy", "mean")))

    merged = (rent_q.merge(vac_q, on="Quarter", how="inner")
                     .sort_values("Quarter")
                     .reset_index(drop=True))
    merged["rent_fwd2_pct"] = merged["eff_rent_psf"].pct_change(2).shift(-2)
    merged = merged.dropna(subset=["rent_fwd2_pct"])

    res = best_split(merged, min_n=min_n)
    res.update({"market": target.title(),
                "rent_q": len(rent_q),
                "vac_q":  len(vac_q),
                "merged_q": len(merged)})

    # ---- plot ----
    if plot:
        plt.figure(figsize=(6,4))
        plt.scatter(merged["Vacancy"], merged["rent_fwd2_pct"])
        if res["thr"] is not None:
            plt.axvline(res["thr"], color="red", ls="--",
                        label=f'best split = {res["thr"]:.1f}%')
        else:
            plt.text(0.05,0.9,"no valid\nsplit",transform=plt.gca().transAxes,color="red")
        plt.title(f'{target.title()}  Class A\nVacancy vs 2-Q Rent Growth')
        plt.xlabel("Availability % (current qtr)")
        plt.ylabel("2-Q Fwd Rent Growth %")
        plt.legend(); plt.tight_layout(); plt.show()

    # console summary
    if res["thr"] is None:
        msg = f"**no breakpoint (need {min_n} obs each side)**"
    else:
        msg = (f"thr {res['thr']:.1f}% | n_low {res['low_n']:2d}  {res['low_mean']:.2%} | "
               f"n_high {res['high_n']:2d}  {res['high_mean']:.2%} | |t| {res['t']:.2f}")
    print(f"{target.title():<12} | {msg}")
    return res


# ================================================================
ALIASES = {
    # ---------- proven working ----------
    "south bay" : ["south bay"],

    # ---------- Downtown L.A. ------------
    "dtla" : [
        "downtown",                      # internal_market_cluster
        "cbd",                           # internal_submarket
        "central business district",
        "los angeles  downtown",        # en dash
        "los angeles - cbd",             # hyphen
        "downtown los angeles",
        "downtown la"
    ]
}
#  Alias-aware analysis wrapper
def analyse_aliases(key, *, min_n=3, min_size=MIN_SIZE, plot=True):
    """
    Tries every alias for `key` until one returns 1 merged quarter.
    Prints which alias matched. Returns stats dict or None.
    """
    for alias in ALIASES.get(key.lower(), [key]):
        stats = analyse_market(alias, min_n=min_n, min_size=min_size, plot=plot)
        if stats["merged_q"] > 0:
            stats["alias_used"] = alias
            return stats
    print(f"No data found for any alias of '{key}'.")
    return None

south_bay_stats = analyse_aliases("south bay")  # South Bay
dtla_stats = analyse_aliases("dtla")     # Downtown L.A.
```
